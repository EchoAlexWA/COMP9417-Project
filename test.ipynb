```python
import os

# 获取目录下所有csv文件
csv_dir = 'data/temporal_features'
csv_files = [f for f in os.listdir(csv_dir) if f.endswith('.csv')]

# 动态读取csv文件并以文件名作为变量名
for csv_file in csv_files:
    var_name = os.path.splitext(csv_file)[0]  # 去掉扩展名作为变量名
    globals()[var_name] = pd.read_csv(os.path.join(csv_dir, csv_file))
    print(f"Loaded {csv_file} into variable '{var_name}'")
```
